```python
import random
# random.random() 함수만으로 로또

# 공을 45 -> 변수 45

balls = list(range(1,46))
print(balls)
```

```python
# balls -> shuffle
for i in range(0,1000000):
    temp = balls[0]
    r_num = int(random.random() * len(balls))   # 0.000 ~ 44.999
    # r_num = math.floor(random.random() * len(balls))
    balls[0] = balls[r_num]
    balls[r_num] = temp

print(balls)
print(balls[:6])
```

```python
# random모듈 안의 sample()함수 사용
# 몇 번 섞었는지 알 수 없음
lotto = random.sample(range(1, 46), 6)
print(lotto)
lotto.sort()
print(lotto)
```

```python
for i in range(10):
    lotto = random.sample(range(1, 46), 6)
    lotto.sort()
    print(f"{i}번째 로또 번호: {lotto}")
```

```python
def my_lotto_create(a):
    result = []
    for i in range(a):
        lotto = random.sample(range(1, 46), 6)
        lotto.sort()
        # print(f"{i+1}번째 로또 번호: {lotto}")
        result.append((tuple(lotto)))
    return tuple(result)
```

```python
result = my_lotto_create(10)
print(result)
```
