## 0. 함수도 객체(값)이다.

```python
def hello():
    print("안녕하세요!")

# 함수를 변수에 저장
f = hello

# 변수로 함수 호출
f()
```

```python
def hello():
    print("안녕하세요!")

def execute(func):
    func()  # 함수 호출 연산자를 사용했다는 것은, func매개변수값이 함수다

execute(hello)
```

```python
# 일반 함수 정의
def add1(a, b):
    return a + b

# 람다 함수
add2 = lambda a, b : a + b # add2에는 "lambda a, b : a + b" 람다가 반환한 함수(객체)가 대입

r1 = add1(5, 1)
print(r1)

r2 = add2(1, 3)
print(r2)
```

```python
# 파이썬에서는 안씀
# 람다를 익명함수라고 부르기도 한다(Java)

r3 = (lambda a, b : a + b)(10,50)
print(r3)
```

```python
def test(fun):
    a = 2
    return fun(2) + 1000

def square(x):
    return x*x

def square2(x):
    return x-x

def square3(x):
    return x*1000


r1 = test(square)
print(r1)

r2 = test(square2)
print(r2)

r3 = test(square3)
print(r3)


# 일일이 하나씩 만들어야 함
# ---------------------

r4 = test(lambda x: x*100)
print(r4)
```

```python
def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True)

print(r)


r2 = sorted(s, key=lambda e:e[1], reverse=False)
print(r2)
```

#### 1) map() - 모든 요소를 변환

```python
nums = [1, 2, 3, 4]

# 리스트의 각 요소(x)에 2를 곱한 결과를 반환
result = map(lambda x: x * 2, nums)

print(result) 
print(list(result))
```

#### 2) filter() - 조건에 맞는 요소만 선택

```python
nums = [1, 2, 3, 4, 5]

# 짝수인 요소만 선택
result = filter(lambda x: x % 2 == 0, nums)

print(result)
print(list(result))
```

#### 3) sorted() - 정렬 기준 지정

```python
names = ["banana", "apple", "kiwi"]

# 문자열의 길이를 기준으로 오름차순 정렬
result = sorted(names, key=lambda x: len(x), reverse=False)

print(result)
```

```python
students = [
    {"이름": "홍길동", "나이": 20, "점수": 90},
    {"이름": "이순신", "나이": 18, "점수": 85},
    {"이름": "김유신", "나이": 22, "점수": 95}
]

# "나이" 값을 기준으로 오름차순 정렬 20,18,22
result = sorted(students, key=lambda x: x["나이"])
print(result)
```

```python
students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수
result = sorted(students, key=lambda x: (x["반"], x["점수"]))
print(result)
```

#### 5) max() - 기준에 따라 최대값 찾기

```python
names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))

print(result) # kiwi
```

#### 6) any() - 하나라도 조건을 만족하는지 확인

```python
nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환

r1 = map(lambda x: x % 2 == 0, nums)
print("r1: ",r1)

r2 = list(r1)
print("r2: ",r2)

result = any(map(lambda x: x % 2 == 0, nums))

print(result) # True
```
