## 1. 변수에 저장하기

```python
def hello():
    print("Hello")

f = hello      # 함수 자체를 저장 (괄호 없음)

f()
```

## 2. 함수의 인수로 전달하기

```python
def square(x):
    return x * x

def test(fun): # 고차함수
    return fun(5)

print(test(square))
```

## 3. 함수를 반환하기

```python
def outer():
    def inner():
        print("inner")

    return inner

f = outer()

f()

outer()()

# 함수 안에 함수 중첩 가능
```

## 4. 자료 구조에 저장하기

```python
def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))
```

```python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

calc = {
    "+": add,
    "-": sub
}

print(calc["+"](10, 5))
print(calc["-"](10, 5))
```

```python
def plus(x):
    return x + 10

a = lambda x: x + 10    # 변수 저장 -> 1급 함수 기능 포함

print(a)
```
