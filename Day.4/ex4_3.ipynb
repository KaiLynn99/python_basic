```python
# 공유 참조

a = 10
b = a

print(a, b) # 10 10
print(a is b)  # True 같은 객체 참조
print(a == b)
```

```python
s1 = "Python"
s2 = "Pytho"

s2 = "Pytho" + "n"

print(s1 is s2) # True
print(s1 == s2) # True
```

```python
s1 = "Python"
s2 = "Pytho"

s2 = "s2" + "n"

print(s1 is s2) 
print(s1 == s2) 
```

```python
nums1 = [10, 20, 30]
nums2 = nums1

print(nums1 is nums2) # True
print(nums1 == nums2) # True

nums1[0] = 999

print(nums1,nums2)
```

```python
nums1 = [10, 20, 30]
nums2 = [10, 20, 30]

print(nums1 is nums2) # False
print(nums1 == nums2) # True

# nums2의 가변데이터는 언제든 바뀔 수 있다
```

```python
a = [1, 2, 3]
b = a

print(a == b) # True
print(a is b) # True

# 내장 함수 id()를 이용하면 동일 객체를 참조하는지 확인 가능
print(id(a))
print(id(b))
```

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b) # True
print(a is b) # False

# 내장 함수 id()를 이용하면 동일 객체를 참조하는지 확인 가능
print(id(a))
print(id(b))
```
