```python
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

```python
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
```

```python
nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)
```

```python
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)
```

```python
nums = [10, 20, 30]
result = nums.pop() # 마지막 인덱스 삭제 후 반환
print(result)
print(nums)
```

```python
nums = [10, 20, 30]
result = nums.pop(1) # 인덱스1 삭제 후 반환
print(result)
print(nums)
```

```python
# clear() : 전체 삭제
nums = [10, 20, 30]
nums.clear()
print(nums)
```

```python
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```

```python
nums = [10, 20, 30, 40]
nums.reverse()
print(nums)
```

```python
nums = [10, 20, 10, 30, 10]
print(nums.count(10)) # 3
```

```python
fruits = ["apple", "banana", "orange"]
print(fruits.index("banana")) # 1
```
