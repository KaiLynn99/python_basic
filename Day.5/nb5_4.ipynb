```python
t = 1, 3, 5 # 팩킹 실행
print(t) 

print(type(t))

l = list(t)
print(l, type(l))

x, y, z = t # 언팩킹 실행
print(x, y, z)


a,b = [3,6] #(3,6) 변경 후 언팩킹 발생
print(a,b)
```

```python
# 패킹, 언패킹 시 * 연산자

# a,b = [1,2,3,4] # too many values to unpack
*a,b = [1,2,3,4]
print(a)
print(b)

print()

c,*d = (1,2,3,4)
print(c)
print(d) 

print()

e,*f,g = (1,2,3,4)
print(e)
print(f)
print(g) 

print()

h = [1,2,3]
print(h)
print(*h) # print(h[0], h[1], h[2])
```
