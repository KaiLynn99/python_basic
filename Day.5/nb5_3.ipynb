```python
a = [1, 2, 3]
b = [4, 5]
result = a + b
print(result) # [1, 2, 3, 4, 5]
print(a)
print(b)
# 원본 변화X

# 리스트 : 가변 
a = a + b
print(a)

# 문자 : 불변
c = "py"
c = c + "thon"
print(c)
```

```python
nums = [10, 20]
result = nums * 3
print(result) # [10, 20, 10, 20, 10, 20]

# [30, 60] NumPy라이브러리로 행렬 객체를 만들어 사용
```

```python
fruits = ["apple", "banana", "orange"]
print("banana" in fruits)

print("grape" in fruits)
print("grape" not in fruits)
```

```python
nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1])

# 인덱스 할당 연산
nums[2] = 99 # 리스트는 가변 객체
print(nums)

print()

n2 = (3,5)
print(n2[0])
n2[0] = 999 # does not support item assignment
```

```python
nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50] 슬라이싱 연산 후에도 리스트 자신은 변경되지 않는다. 

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[1::2])

nums = [10, 20, 30, 40, 50]
print(nums[::-1])

print(nums)
```

```python
a = [10, 20, 30]
b = a

print(a)
print(b)
print(a == b)
print(a is b)

b[0] = 999
print(a) 
print(b) 
```

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b) # True
print(a is b) # False

# 각각의 값임
```

```python
nums = [10, 20, 30, 40]

print(len(nums))
nums = [10, 20, 30, 40]

print(max(nums))
print(min(nums))
print(sum(nums))
```

```python
# append() : 맨 뒤에 1개 추가
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

```python
# insert() : 원하는 위치에 추가
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
```

```python
nums = [50, 10, 40, 20]
result = sorted(nums)
print(nums)
print(result)
```

```python
# 정렬 메서드
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)
```
