```python
# 1. 정수 (int)
a = 10
print(a)
print(type(a))
```

```python
# 2. 부동 소수점 수 (float) : 오차를 가진다
b = 3.14
print(b)
print(type(b))

print(0.1+0.2)
```

```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))
print(3>5)
```

```python
# 5. 문자열 (str)
e = "Python"
print(e)
print(type(e))

print(e[1])

# e[1] = "k" # error 문자열 객체는 불변이다
```

```python
# 6. 리스트 (list)
f = [1, 2, 3]
print(f)
print(type(f))

print(f[1])

f[1] = "abc"
print(f)
```

```python
# 7. 튜플 (tuple)
g = (1, 2, 3)
print(g)
print(type(g))

print(g[1])

# g[1] = 9
# print(g)
# e[1] = 9 # error 튜플은 불변이다
```

```python
# 8. 사전 딕셔너리 (dict)
h = {"id": "admin", "pw" : "1234" }
print(h)
print(type(h))

print(h["id"])

h["pw"] = "0000"
print(h)
```

```python
# 9. 집합 (set)
i = {1, 2, 3}
print(i)
print(type(i))

# print(i[0]) X
# 일부분 출력 불가
# + 중복값 허용 X
```

```python
# 10. 범위 (range)
j = range(0,5)
print(j)
print(type(j))
print(list(j))
```
