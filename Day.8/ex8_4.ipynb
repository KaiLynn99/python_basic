```python
# 1) 결측치 확인

import csv

with open("score.csv", "r", encoding="utf-8") as f:
    scores = list(csv.DictReader(f))


missing_age_count = 0
missing_korean_count = 0
missing_english_count = 0

missing_values = ["", "결측", "NaN"]

for student in scores:
    if student["나이"] in missing_values: 
        missing_age_count += 1

    if student["국어"] in missing_values: 
        missing_korean_count += 1
    
    if student["영어"] in missing_values:
        missing_english_count += 1

print("'나이' 결측치 개수:", missing_age_count)
print("'국어' 결측치 개수:", missing_korean_count)
print("'영어' 결측치 개수:", missing_english_count)
```

```python
import csv

with open("score.csv", "r", encoding="utf-8") as f:
    scores = list(csv.DictReader(f))

empty_count = 0
missing_text_count = 0
nan_count = 0

for student in scores:
    for column in ["나이", "국어", "영어"]:
        value = student[column]

        if value == "":
            empty_count += 1
        if value == "결측":
            missing_text_count += 1
        if value == "NaN":
            nan_count += 1

print('[""]',"결측치 개수:", empty_count)
print("[결측] 결측치 개수:",missing_text_count)
print("[Nan] 결측치 개수:", nan_count)
```
