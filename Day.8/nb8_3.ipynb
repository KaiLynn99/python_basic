```python
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip()) # line에 양쪽 공백제거(줄바꿈포함)하여 print함수의 기본 줄바꿈만으로 출력
```

```python
# 한 라인을 구분자","를 이용하여 두개의 문자열 값으로 저장

names = []
scores = []

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        # print(name, int(score))
        names.append(name)
        scores.append(score)

print(names)
print(scores)
```

```python
s = [] #[(),(),()]

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        a = line.strip().split(",")
        # print(name, int(score))
        s.append(tuple(a))

print(s)
```

```python
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    # reader() : csv모듈 안에 있는 함수
    reader = csv.reader(f)
    print(type(reader)) # <class '_csv.reader'>
    for row in reader:
        print(row)
        print(type(row)) # <class 'list'>
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    next(reader) # -> 첫번째 찾고 그 다음 연결 (n번째 X)

    for row in reader:
        print(row)
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # 헤드라인 키, 한줄의 내용의 값

    for row in reader:
        print(row)
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader) # -> r는 읽기 전용이므로, list로 바꿔서 수정 가능 

print(data)

data[2]["영어"] = 80


print(data[2])
```

```python
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())

    # 수정

    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")
```
